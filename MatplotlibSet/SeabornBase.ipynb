```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    #fig = plt.figure(figsize=(8,6))
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7- i) * flip)
sinplot()
```

```python
sns.set()
fig = plt.figure(figsize=(8,6))
sinplot()
```
